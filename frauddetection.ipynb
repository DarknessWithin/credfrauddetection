```python
data1 = data[data['Class'] == 1]
```

```python
data2 = data[data['Class'] == 0]
```

```python
data3 = data2.sample(n=492,random_state=42)
```

```python
dataf = pd.concat([data1,data3])
```

```python
shuffled_df = dataf.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
shuffled_df.head()
```

```python
X = shuffled_df.drop(['Class'], axis=1)
```

```python
Y = shuffled_df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaler.transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor = LogisticRegression(penalty='l1', solver='liblinear')


```

```python
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```
